```python
mu, sigma = 2., 1. # mean and standard deviation
s = np.random.lognormal(mu, sigma, 100000)
print(s[0:10])
```

```python
count, bins, ignored = plt.hist(s, 100, density=True, align='mid',range=(0,250))
x = np.linspace(min(bins), max(bins), 10000)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))
plt.plot(x, pdf, linewidth=2, color='r')
plt.axis('tight')
plt.show()
```

```python
q = np.random.random(size=100000)
offset = mu + sigma**2
scale = -np.sqrt(2)*sigma
y = np.exp(offset + scale*erfcinv(2*q))
#print(y)
```
